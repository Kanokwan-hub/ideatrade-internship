```python
import pandas as pd

def trailing_stop(df_buy, stop_pct=0.05, sleep_days=5):
    exit_days = []
    for i, row in df_buy.iterrows():
        if row['action'].lower() != 'buy':
            exit_days.append(None)
            continue

        entry_price = row['cs']
        max_price = entry_price
        exit_day = 15

        for day in [5, 10, 15]:
            if day < sleep_days:
                continue
            cs_col = f'cs{day}'
            if pd.notna(row.get(cs_col)):
                price = row[cs_col]
                max_price = max(max_price, price)
                if price < max_price * (1 - stop_pct):
                    exit_day = day
                    break

        exit_days.append(exit_day)

    df_result = df_buy.copy()
    df_result['exit_trailing'] = exit_days
    return df_result

df_buy = trailing_stop(df_buy)

pred_trailing = []
for i in range(len(df_buy)):
    exit_day = df_buy.iloc[i]['exit_trailing']
    if pd.isna(exit_day):
        pred_trailing.append(False)
        continue

    entry = df_buy.iloc[i]['cs']
    cs_col = f'cs{int(exit_day)}'

    if cs_col in df_buy.columns and pd.notna(df_buy.iloc[i][cs_col]):
        future = df_buy.iloc[i][cs_col]
        gain = (future - entry) / entry
        pred_trailing.append(gain >= 0.05)
    else:
        pred_trailing.append(False)

df_buy['predict_trailing'] = pred_trailing


pred_cs15 = []
for i in range(len(df_buy)):
    entry = df_buy.iloc[i]['cs']
    if 'cs15' in df_buy.columns and pd.notna(df_buy.iloc[i]['cs15']):
        future = df_buy.iloc[i]['cs15']
        gain = (future - entry) / entry
        pred_cs15.append(gain >= 0.075)
    else:
        pred_cs15.append(False)

df_buy['predict_cs15'] = pred_cs15
df_buy['predict_combined'] = df_buy['predict_trailing'] | df_buy['predict_cs15']

list_analys = df_buy['analys'].unique().tolist()
list_true = []
list_false = []

for a in list_analys:
    df_a = df_buy[df_buy['analys'] == a].reset_index(drop=True)
    no_true = df_a[df_a['predict_combined'] == True].shape[0]
    no_false = df_a[df_a['predict_combined'] == False].shape[0]
    list_true.append(no_true)
    list_false.append(no_false)

df_buy_analys = pd.DataFrame({
    'analys': list_analys,
    'true': list_true,
    'false': list_false
})
df_buy_analys['total'] = df_buy_analys['true'] + df_buy_analys['false']
df_buy_analys['accuracy'] = df_buy_analys['true'] / df_buy_analys['total']

print("TRUE:", df_buy['predict_combined'].sum())
print("FALSE:", (~df_buy['predict_combined']).sum())

df_buy_analys
```

```python
import pandas as pd
from datetime import datetime

df_buy['date'] = pd.to_datetime(df_buy['date'], format='mixed')

def sell_signal_exit(df, sleep_days=5):
    exit_days = []

    for i, row in df.iterrows():
        if row['action'].lower() != 'buy':
            exit_days.append(None)
            continue

        stock = row['stock_name']
        entry_date = row['date']

        mask = (
            (df['stock_name'] == stock) &
            (df['action'].str.lower() == 'sell') &
            (pd.to_datetime(df['date']) >= entry_date + pd.Timedelta(days=sleep_days))
        )

        if df[mask].empty:
            exit_days.append(None)
        else:
            sell_date = df[mask]['date'].min() 
            delta_days = (sell_date - entry_date).days
            if delta_days <= 5:
                exit_day = 5
            elif delta_days <= 10:
                exit_day = 10
            else:
                exit_day = 15
            exit_days.append(exit_day)

    df_result = df.copy()
    df_result['exit_sell_signal'] = exit_days
    return df_result

df_buy = sell_signal_exit(df_buy)

pred_sell_signal = []
for i in range(len(df_buy)):
    exit_day = df_buy.iloc[i]['exit_sell_signal']
    if pd.isna(exit_day):
        pred_sell_signal.append(False)
        continue

    entry = df_buy.iloc[i]['cs']
    cs_col = f'cs{int(exit_day)}'

    if cs_col in df_buy.columns and pd.notna(df_buy.iloc[i][cs_col]):
        future = df_buy.iloc[i][cs_col]
        gain = (future - entry) / entry
        pred_sell_signal.append(gain >= 0.05)
    else:
        pred_sell_signal.append(False)

df_buy['predict_sell_signal'] = pred_sell_signal

pred_cs15 = []
for i in range(len(df_buy)):
    entry = df_buy.iloc[i]['cs']
    if 'cs15' in df_buy.columns and pd.notna(df_buy.iloc[i]['cs15']):
        future = df_buy.iloc[i]['cs15']
        gain = (future - entry) / entry
        pred_cs15.append(gain >= 0.075)
    else:
        pred_cs15.append(False)

df_buy['predict_cs15'] = pred_cs15
df_buy['predict_combined'] = df_buy['predict_sell_signal'] | df_buy['predict_cs15']

list_analys = df_buy['analys'].unique().tolist()
list_true = []
list_false = []

for a in list_analys:
    df_a = df_buy[df_buy['analys'] == a].reset_index(drop=True)
    no_true = df_a[df_a['predict_combined'] == True].shape[0]
    no_false = df_a[df_a['predict_combined'] == False].shape[0]
    list_true.append(no_true)
    list_false.append(no_false)

df_buy_analys = pd.DataFrame({
    'analys': list_analys,
    'true': list_true,
    'false': list_false
})
df_buy_analys['total'] = df_buy_analys['true'] + df_buy_analys['false']
df_buy_analys['accuracy'] = df_buy_analys['true'] / df_buy_analys['total']

print("TRUE:", df_buy['predict_combined'].sum())
print("FALSE:", (~df_buy['predict_combined']).sum())

df_buy_analys
```

```python
import pandas as pd
from datetime import timedelta

def combined_exit_strategy(df, stop_pct=0.05, sleep_days=5):
    ts_flags = []
    ss_flags = []
    cs15_flags = []

    for i, row in df.iterrows():
        if row['action'].lower() != 'buy':
            ts_flags.append(False)
            ss_flags.append(False)
            cs15_flags.append(False)
            continue

        entry_price = row['cs']
        entry_date = pd.to_datetime(row['date'])

        # Trailing Stop Exit (after sleep_days)
        max_price = entry_price
        ts_triggered = False
        for day in [5, 10, 15]:
            if day < sleep_days:
                continue
            col = f'cs{day}'
            if pd.notna(row.get(col)):
                price = row[col]
                max_price = max(max_price, price)
                if price < max_price * (1 - stop_pct):
                    ts_triggered = True
                    break
        ts_flags.append(ts_triggered)

        # Sell Signal with gain ≥ 5%
        ss_triggered = False
        df_sell = df[
            (df['stock_name'] == row['stock_name']) &
            (df['action'].str.lower() == 'sell') &
            (pd.to_datetime(df['date']) >= entry_date + timedelta(days=sleep_days))
        ]
        if not df_sell.empty:
            first_sell_date = pd.to_datetime(df_sell['date'].min())
            if 'cs' in df_sell.columns:
                price_sell = df_sell[df_sell['date'] == first_sell_date]['cs'].values[0]
                gain = (price_sell - entry_price) / entry_price
                if gain >= 0.05:
                    ss_triggered = True
        ss_flags.append(ss_triggered)

        # cs15 gain ≥ 7.5%
        if pd.notna(row.get('cs15')):
            gain15 = (row['cs15'] - entry_price) / entry_price
            cs15_flags.append(gain15 >= 0.03)
        else:
            cs15_flags.append(False)

    df_result = df.copy()
    df_result['ts_flag'] = ts_flags
    df_result['ss_flag'] = ss_flags
    df_result['cs15_flag'] = cs15_flags

    # Final prediction:
    # ต้องมี ts_flag หรือ ss_flag ก่อน (อย่างน้อยหนึ่ง True)
    # และ cs15_flag ต้องเป็น True ด้วย
    df_result['predict_combined'] = ((df_result['ts_flag'] | df_result['ss_flag']) | df_result['cs15_flag'])

    return df_result

# Run strategy
df_buy = combined_exit_strategy(df_buy)

# Accuracy summary (เหมือนเดิม)
list_analys = df_buy['analys'].unique().tolist()
list_true = []
list_false = []

for a in list_analys:
    df_a = df_buy[df_buy['analys'] == a].reset_index(drop=True)
    no_true = df_a[df_a['predict_combined'] == True].shape[0]
    no_false = df_a[df_a['predict_combined'] == False].shape[0]
    list_true.append(no_true)
    list_false.append(no_false)

df_buy_analys = pd.DataFrame({
    'analys': list_analys,
    'true': list_true,
    'false': list_false
})
df_buy_analys['total'] = df_buy_analys['true'] + df_buy_analys['false']
df_buy_analys['accuracy'] = df_buy_analys['true'] / df_buy_analys['total']

# Output summary
print("TRUE:", df_buy['predict_combined'].sum())
print("FALSE:", (~df_buy['predict_combined']).sum())

df_buy_analys

```

```python
# 1. Accuracy ต่อหุ้น แต่ละ Analyst
def accuracy_per_stock_analyst(df, min_preds=3):
    rows = []

    for stock in df['stock_name'].unique():
        df_stock = df[df['stock_name'] == stock]
        for analyst in df_stock['analys'].unique():
            df_analyst = df_stock[df_stock['analys'] == analyst]
            num_preds = df_analyst.shape[0]
            num_correct = df_analyst[df_analyst['predict_combined'] == True].shape[0]
            accuracy = num_correct / num_preds if num_preds > 0 else None

            rows.append([
                stock,
                analyst,
                num_preds,
                num_correct,
                round(accuracy, 4) if accuracy is not None else None
            ])

    accuracy_by_stock_analyst = pd.DataFrame(
        rows,
        columns=['stock_name', 'analyst', 'num_predictions', 'num_correct', 'accuracy']
    )

    filtered_df = accuracy_by_stock_analyst[accuracy_by_stock_analyst['num_predictions'] >= min_preds]
    filtered_df = filtered_df.sort_values(by='accuracy', ascending=False).reset_index(drop=True)

    print(f"มีทั้งหมด {len(filtered_df)} ที่ pred >= {min_preds} ครั้ง")
    return filtered_df

df_result = accuracy_per_stock_analyst(df_buy, min_preds=3)
display(df_result)
```

```python
def compute_overall_analyst_accuracy():
    rows = []
    for stock in df_buy['stock_name'].unique():
        df_stock = df_buy[df_buy['stock_name'] == stock]
        for analyst in df_stock['analys'].unique():
            df_analyst = df_stock[df_stock['analys'] == analyst]
            num_preds = df_analyst.shape[0]
            num_correct = df_analyst[df_analyst['predict_combined'] == True].shape[0]
            rows.append([
                stock,
                analyst,
                num_preds,
                num_correct
            ])

    df_all = pd.DataFrame(
        rows,
        columns=['stock_name', 'analyst', 'num_predictions', 'num_correct']
    )
    df_agg = df_all.groupby('analyst').agg({
        'num_predictions': 'sum',
        'num_correct': 'sum'
    }).reset_index()

    df_agg['accuracy'] = (df_agg['num_correct'] / df_agg['num_predictions']).round(4)
    df_agg = df_agg.sort_values(by='accuracy', ascending=False).reset_index(drop=True)
    return df_agg

# เรียกฟังก์ชันเก็บผลลัพธ์ไว้ตัวแปร
df_analyst_accuracy = compute_overall_analyst_accuracy()

def get_analyst_accuracy(analyst_name=None):
    if analyst_name is None:
        # ถ้าไม่ใส่ชื่อ analyst, แสดงทั้งหมด
        return df_analyst_accuracy
    else:
        # ถ้าใส่ชื่อ analyst, กรองแค่ชื่อที่ตรงกัน (case-insensitive)
        return df_analyst_accuracy[df_analyst_accuracy['analyst'].str.lower() == analyst_name.lower()]

# ตัวอย่างเรียกดูทั้งหมด
display(get_analyst_accuracy())
```

```python
future_close_cols = ['cs5','cs10','cs15']

def calculate_holding_days(row):
    day_close = row['cs']
    for day, col in enumerate(future_close_cols, start=1):
        future_price = row[col]
        if pd.isna(future_price):
            continue
        # ถ้ากำไรเกิน 3%
        if (future_price - day_close) / day_close >= 0.03:
            return day  
    return None  

df_buy['holding_days'] = df_buy.apply(calculate_holding_days, axis=1)

rows = []
for stock in df_buy['stock_name'].unique():
    df_stock = df_buy[df_buy['stock_name'] == stock]
    for analyst in df_stock['analys'].unique():
        df_analyst = df_stock[df_stock['analys'] == analyst]
        count_trades = df_analyst.shape[0]
        avg_holding = df_analyst['holding_days'].dropna().mean() if not df_analyst['holding_days'].dropna().empty else None
        rows.append([stock, analyst, count_trades, round(avg_holding, 2) if avg_holding is not None else None])

df_avg_holding = pd.DataFrame(rows, columns=['stock_name', 'analyst', 'num_trades', 'avg_holding_days'])

def get_avg_holding_days(stock_name=None, analyst=None):
    df_filtered = df_avg_holding
    if stock_name is not None:
        df_filtered = df_filtered[df_filtered['stock_name'] == stock_name]
    if analyst is not None:
        df_filtered = df_filtered[df_filtered['analyst'] == analyst]
    if df_filtered.empty:
        print("ไม่พบข้อมูลที่ตรงกับเงื่อนไข")
        return None
    return df_filtered

print("เฉลี่ยถือหุ้นของ Dome Kunprayoonsawad:")
display(get_avg_holding_days(analyst='Dome Kunprayoonsawad'))

print("เฉลี่ยถือหุ้นของหุ้น AMATA:")
display(get_avg_holding_days(stock_name='AMATA'))

print("เฉลี่ยถือหุ้นของ Dome Kunprayoonsawad กับหุ้น AMATA:")
display(get_avg_holding_days(stock_name='AMATA', analyst='Dome Kunprayoonsawad'))
```

```python
import json

def get_analysts_for_stock(specific_stock, detailed_df = detailed_df):
    # specific_stock = "HMPRO"  

    df_filtered = detailed_df[detailed_df['stock_name'] == specific_stock]

    if df_filtered.empty:
        output = {
            "error": f"Stock name '{specific_stock}' not found in the dataset."
        }
    else:
        analyst_details = []
        for i in range(1, int(df_filtered['number_of_analysts'].values[0]) + 1):
            analyst_name = df_filtered[f'analys#{i}_name'].values[0]
            num_predictions = int(df_filtered[f'analys#{i}_num_predictions'].values[0])  
            accuracy = float(df_filtered[f'analys#{i}_accuracy'].values[0]) 
            analyst_details.append({
                "analyst_name": analyst_name,
                "num_predictions": num_predictions,
                "accuracy": accuracy
            })

        analysts = sorted(analyst_details, key=lambda x: (x['accuracy'], x['num_predictions']), reverse=True)

        output = {
            "stock_name": specific_stock,
            "num_analysts": int(df_filtered['number_of_analysts'].values[0]), 
            "analysts": analysts
        }

    # with open(f"{specific_stock}_top_5_analysts.json", "w", encoding="utf-8") as json_file:
    #     json.dump(output, json_file, ensure_ascii=False, indent=4)

    print(json.dumps(output, ensure_ascii=False, indent=4)) 
    return output
```
